# KMeans

Outliers should not be present in kmeans algorithm

# Transform Outliers

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler ,Normalizer
```

```python
df['N'].var(), df['N'].mean()
```

```python
# Nomalizing outliers between 0 to 1

std = StandardScaler()

x_transformed = std.fit_transform(x)
```

```python
x_transformed
```

```python
x.shape , x_transformed.shape  # only difference is that x is in dataframe and x_transformed in array
```

```python
from sklearn.cluster import KMeans
# KMeans()
# euclidian distance method

# mera bhai mujhse 90% similar ==> 10% different
# uska bhai mujhse 90% different ==> 10% similar
# 12 points should be <= 12 groups (by default 8 cluster group made by kmeans)

kmeans = KMeans()

kmeans.fit(x_transformed)
```

# Finding Grp Numbers formed by KMeans()

```python
kmeans.labels_
```

```python
np.unique(kmeans.labels_)  # group numbering 0 to 7 total 8
```

```python
kmeans.labels_.shape , df.shape
```

```python
df['cluster_no'] = kmeans.labels_
```

```python
df.head()
```

```python
group_five = df[df['cluster_no']==5]
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_transformed,kmeans.labels_)  # silhouette_score range is -1 to +1 where 0 means neutral and if model score approach toward 1 means model is good 
```
